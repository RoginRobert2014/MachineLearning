```python
import pandas as pd
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn import svm
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn import metrics
```

```python
dataset = pd.read_csv("CE802_P2_Data.csv")
dataset.head()
```

```python
datasetMean = dataset.fillna(dataset.mean())
# Experimenting without F15 column 
# X= datasetMean.loc[:, :'F14']

X= datasetMean.loc[:, :'F15']
Y= datasetMean['Class']

# Experimenting with median value
# datasetMedian = dataset.fillna(dataset.median()) 
# X= datasetMedian.loc[:, :'F15']
# Y= datasetMedian['Class'] 
```

#  a) Investigate the performance of decision tree classifier

```python
#split data into training and test set
X_train_DT, X_test_DT, y_train_DT, y_test_DT = train_test_split(X, Y, test_size = 0.10, random_state= 105)

#Create decision tree classifier object
decTreeClassObj = DecisionTreeClassifier(random_state=0)

#Training deciosion tree classifier
decTreeClassObj = decTreeClassObj.fit(X_train_DT, y_train_DT)

#predict the result for test dataset
y_Prediction_DT = decTreeClassObj.predict(X_test_DT)

# Model Performance, how often is the classifier correct
print("Accuracy:", metrics.accuracy_score(y_test_DT, y_Prediction_DT))

print("Precision:", metrics.precision_score(y_test_DT, y_Prediction_DT))

print("Recall:", metrics.recall_score(y_test_DT, y_Prediction_DT))
```

# a) Investigate the performance of Random Forest Classifier

```python
#split data into training and test set
X_train_RFC, X_test_RFC, y_train_RFC, y_test_RFC = train_test_split(X, Y, test_size = 0.1, random_state= 105)

#Create a Gaussian classifier
rfClassObj = RandomForestClassifier(n_estimators=200, random_state=0)

#Training deciosion tree classifier
rfClassObj = rfClassObj.fit(X_train_RFC, y_train_RFC)

#predict the result for test dataset
y_Prediction_RFC = rfClassObj.predict(X_test_RFC)

# Model Accuracy, how often is the classifier correct
print("Accuracy:", metrics.accuracy_score(y_test_RFC, y_Prediction_RFC))

print("Precision:", metrics.precision_score(y_test_RFC, y_Prediction_RFC))

print("Recall:", metrics.recall_score(y_test_RFC, y_Prediction_RFC))
```

# a) Investigate the performance of svm classifier

```python
#split data into training and test set
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.1, random_state= 105)

#Create a svm classifier
svmClassObj = svm.SVC(kernel='linear') #using linear Kernel

#Training svm classifier using training sets
svmClassObj.fit(X_train, y_train)

#predict the result for test dataset
y_Prediction = svmClassObj.predict(X_test)

# Model Accuracy
print("Accuracy:",metrics.accuracy_score(y_test, y_Prediction))

print("Precision:", metrics.precision_score(y_test, y_Prediction))

print("Recall:", metrics.recall_score(y_test, y_Prediction))
```

# b) Prediction on a hold-out test set using Random Forest

```python
datasetTest = pd.read_csv("CE802_P2_Test.csv")

datasetMeanTest = datasetTest.fillna(datasetTest.mean())

X_Testset= datasetMeanTest.loc[:, :'F15']

#split data into training and test set
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.1, random_state= 105)

# Create a Random Forest Classifier object
rfClassObj = RandomForestClassifier(n_estimators=200, random_state=0)

#Training Random Forest Classifier using training sets
rfClassObj.fit(X_train, y_train)

#predict the result for test dataset
y_Prediction = rfClassObj.predict(X_Testset)

datasetMeanTest['Class'] = y_Prediction


#Save the new results in a new file CE802_P2_Test.csv
datasetMeanTest.to_csv("CE802_P2_Test_V2.csv", index=False)
datasetMeanTest
```
