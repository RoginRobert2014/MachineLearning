```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

# 3 a) Investigate the performance using machine learning procedures

```python
dataset = pd.read_csv("CE802_P3_Data.csv")
```

```python
#Change the string values in column F5 and F6 to numerical values
datasetF5 = pd.factorize(dataset['F5'])
dataset['F5'] = datasetF5[0]
datasetF6= pd.factorize(dataset['F6'])
dataset['F6'] = datasetF6[0]

dataset.head()
```

```python
X = dataset.loc[:, :'F16']
Y= dataset['Target']
```

# 3 a) Investigate the performance using Linear Regression

```python
#split data into training and test set
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.1, random_state= 0)

#Create linear regression object
linRegClassObj = LinearRegression()

#Train the model using training sets
linRegClassObj.fit(X_train, y_train)

#predict the result for test dataset
y_Prediction = linRegClassObj.predict(X_test)

# The mean squared error
print('Mean Squared Error: ', mean_squared_error(y_test, y_Prediction))

# The mean absolute error
print('Mean Absolute Error: ', mean_absolute_error(y_test, y_Prediction))

# RMSE for Linear Regression
print('RMSE for Linear Regression', np.sqrt(mean_squared_error(y_test, y_Prediction)))

# The coefficient of determination 
print('Coefficient of determination: ', r2_score(y_test, y_Prediction))
```

# 3 a) Investigate the performance using Random Forest Regression

```python
#split data into training and test set
X_train_RFR, X_test_RFR, y_train_RFR, y_test_RFR = train_test_split(X, Y, test_size = 0.1, random_state= 1)

#Create random forest object
regr = RandomForestRegressor(max_depth=15, random_state=115)

#Train the model using training sets
regr.fit(X_train_RFR, y_train_RFR)

#predict the result for test dataset
y_Prediction_RFR = regr.predict(X_test_RFR)

# The mean squared error
print('Mean Squared Error: ', mean_squared_error(y_test_RFR, y_Prediction_RFR))

# The mean absolute error
print('Mean absolute error: ', mean_absolute_error(y_test_RFR, y_Prediction_RFR))

# RMSE for Random Forest Regression
print('RMSE for Random Forest Regression', np.sqrt(mean_squared_error(y_test_RFR, y_Prediction_RFR)))

# The coefficient of determination 
print('Coefficient of determination: ', r2_score(y_test_RFR, y_Prediction_RFR))
```

# 3 a) Investigate the performance using Support Vector Regression

```python
#split data into training and test set
X_train_SVR, X_test_SVR, y_train_SVR, y_test_SVR = train_test_split(X, Y, test_size = 0.1, random_state= 120)
svRegr = make_pipeline(StandardScaler(), SVR(C=1190.0, epsilon=0.1))

# #Train the model using training sets
svRegr.fit(X_train_SVR, y_train_SVR)

# #predict the result for test dataset
y_Prediction_SVR = svRegr.predict(X_test_SVR)

# The mean squared error
print('Mean Squared Error: ', mean_squared_error(y_test_SVR, y_Prediction_SVR))

# The mean absolute error
print('Mean absolute error: ', mean_absolute_error(y_test_SVR, y_Prediction_SVR))

# RMSE for Support Vector Machine Regression
print('RMSE for Support Vector Machine Regression', np.sqrt(mean_squared_error(y_test_SVR, y_Prediction_SVR)))

# The coefficient of determination 
print('Coefficient of determination: ', r2_score(y_test_SVR, y_Prediction_SVR))
```
